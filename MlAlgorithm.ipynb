# Fill empty values with mean

```python
mean_review_rating = data['review_scores_rating'].mean()
mean_review_accuracy = data['review_scores_accuracy'].mean()
mean_review_cleanliness = data['review_scores_cleanliness'].mean()
mean_review_checkin = data['review_scores_checkin'].mean()
mean_review_communication = data['review_scores_communication'].mean()
mean_review_location = data['review_scores_location'].mean()
mean_review_value = data['review_scores_value'].mean()
```

```python
data['review_scores_rating'].fillna(value=mean_review_rating,inplace=True)
data['review_scores_accuracy'].fillna(value=mean_review_accuracy,inplace=True)
data['review_scores_cleanliness'].fillna(value=mean_review_cleanliness,inplace=True)
data['review_scores_checkin'].fillna(value=mean_review_checkin,inplace=True)
data['review_scores_communication'].fillna(value=mean_review_communication,inplace=True)
data['review_scores_location'].fillna(value=mean_review_location,inplace=True)
data['review_scores_value'].fillna(value=mean_review_value,inplace=True)
data['bathrooms'].fillna(0,inplace=True)
```

```python
empty_cols = data.columns[data.isna().any()].tolist()
# data.fillna(0,inplace=True)
empty_cols
```

```python
for col in empty_cols:
    mean = data[col].mean()
    data[col].fillna(mean,inplace=True)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv("listing_review_cleaned.csv")
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['host_total_listings_count','accommodates','bedrooms','beds','price','minimum_nights','maximum_nights','minimum_nights_avg_ntm','maximum_nights_avg_ntm','availability_30','availability_60','availability_90','availability_365','number_of_reviews','host_since_days','number_of_amenities','calculated_host_listings_count','calculated_host_listings_count_entire_homes','calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms']
```

```python
scale = StandardScaler()
features_df = data[cols]
type(features_df)
```

```python
gg_scaled.head()
```
